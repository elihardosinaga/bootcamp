```python
# Operator Aritmatika

angka1 = int(input("Masukkan angka 1 = "))
angka2 = int(input("Masukkan angka2 = "))

print("Tambah : ", angka1 + angka2)
print("kurang : ", angka1 -angka2)
print("Bagi : ", angka1 / angka2)
print("kali : ", angka1 * angka2)
print("Sisa bagi : ", angka1 % angka2)
```

```python
# operator penugasan

angka1 = 100
print(angka1)
angka1 += 20
print(angka1)
angka1 /= 3
print(angka1)
angka1 *= 2
print(angka1)
angka1 %= 4
print(angka1)
```

```python
# operator perbandingan

angka1 = 100
angka2 = 200

print("Lebih besar :", angka1 > angka2)
print("lebih kecil :", angka1 < angka2)
print("Sama dengan :", angka1 == angka2)
print("Tidak sama dengan : ", angka1 != angka2)

```

```python
# operator logika

angka1 = 100
angka2 = 200
angka3 = 300

# and = salah satu kondisi salah (False) maka hasilnya salah (False)
# or = salah satu kondisi benar (True) maka hasilnya benar (True)
# not / negasi = membalikkan hasil kondisi awal

print("Hasil and : ", angka1 < angka2 and angka3 < angka1)   # true dan false = false
print("Hasil or : ", angka1 > angka3 or angka2 < angka3)     # false dan true = true
print("Hasil not dan and : ", not(angka1 > angka2 and angka3 < angka1))   # true dan false = true

```

```python
## if statement


username = input("Masukkan username : ")
password = input("Masukkan password : ")


if username == "admin" and password == "123" :
  print("selamat datang admin")
elif username == "user" and password == "321":
  print("selamat datang user")
elif username == "boos" and password == "213":
  print("selamat datang Boss")
else :
  print("username atau password salah")
```

```python
import pandas as pd
import numpy as np

dict = {
    "I" : [100, 90, np.nan, 95],
    "II" : [34, 43, 55, np.nan],
    "III" : [np.nan, 23, 32, 44]
}

df = pd.DataFrame(dict)
print(df)

df = df.fillna(df.mean())
print(df)
```

```python
df = pd.DataFrame(dict)
print(df)

df = df.interpolate()    # mengisi nilai data yang hilang dengan menghitung nilai diantara nilai kedua yang ada dianataranya
                        # dan biasanya digunakan untuk menhitung nilai suhu, harga, sensor, waktu dan lainlainnya.

print(df)
```

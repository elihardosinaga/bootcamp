```python
# looping (for, while)

# for = perulangan yang sudah diketahui berulang kali

for angka in range(5): # range (start)
  print(angka)
  print("Halo world")
```

```python
for angka in range(1,5):
  print(angka)
  print("HALO WORLD")

```

```python
for angka in range(1, 10, 3):
  print(angka)
```

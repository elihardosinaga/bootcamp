```python
#  basic statistic

data = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/penyakit_jantung.csv")
print(data)
```

```python
!pip install scipy
```

```python
import numpy as np
from scipy import stats

thalach = data["thalach"]
print(thalach)
```

```python
# mencari rata rata

mean = np.mean(thalach)
print(mean)

print(thalach.mean())
```

```python
# mencari nilai tengah (median)

median = np.median(thalach)
print(median)
```

```python
print(thalach.median())
```

```python
# mencari nilai terbanyak yang muncul (modus / mode)

mode = stats.mode(thalach)
print(mode)
print(mode[0])
print(mode[1])
```

```python
# varian (mengukur sejauh mana data tersebar dari rata rata )

variance = np.var(thalach)
print(variance)
```

```python
print(thalach.var())
```

```python
# standar deviasi ( mengukur persebaran data )

std = np.std(thalach)
print(std)
print(thalach.mean())
```

```python
# batas atas : 149 + 22 = 171
# rata rata : 100
# batas bawah : 149 - 22 = 127
# nilai yang diluar dari batas / bawah = (outlier)
```

```python
# skweness (melihat dominan data)
# jika hasil plus = data dominan kurang dari rata rata
# jika hasil minus = data dominan lebih besar dari rata rata

skewness = stats.skew(thalach)
print(skewness)
```

```python
from numpy._core.defchararray import lower
# kuartil (data terbagi menjadi beberapa bagian)
#tujuannya untuk mencari outlier
# perbedaan dari std, kuartil menggunakan nilai tengah

data = thalach.sort_values()

q1 = np.percentile(data, 25)
q2 = np.percentile(data, 50)
q3 = np.percentile(data, 75)

iqr = q3 - q1

lower = q1 - (1.5 * iqr)
upper = q3 + (1.5 * iqr)

outlier = thalach.loc[(thalach > upper) | (thalach < lower)]
thalach = thalach.drop(outlier.index)
print(thalach)
```

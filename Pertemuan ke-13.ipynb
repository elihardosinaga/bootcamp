```python
import pandas as pd

data = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/penyakit_jantung.csv")
print(data)
```

```python
age = data.groupby("age")["age"].count()
print(age)
```

```python
plt.figure(figsize = (8,5))
plt.plot(age.index, age.values, color = "green", marker = "o", ls = "--")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.yticks([0, 5, 10, 15, 20])
plt.show()
```

```python
# bar plot (grafik batang)

plt.figure(figsize = (8,5))
plt.bar(age.index, age.values, color = "green")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
# scatter plot (grafik persebaran)

plt.figure(figsize = (8,5))
plt.scatter(age.index, age.values, color = "green")
plt.show()
```

```python
# histogram (grafik frekuensi)

plt.figure(figsize = (8,5))
plt.hist(age.index, color = "green", bins = 5)
plt.title("Age Distribution")
plt.xlabel("Nilai")
plt.ylabel("Frekuensi")
plt.show()
```

```python
# grafik pie

gender = data.groupby("sex")["sex"].count()
print(gender)

plt.pie(gender.values, labels = gender.index, autopct = "%1.2f%%")
plt.legend()
plt.show()
```

```python
# subplot

plt.figure(figsize = (10,5))

plt.subplot(121) # subplot(baris, kolom, letak plot )
plt.plot(age.index, age.values, color = "green", marker = "o", ls = "--")
plt.title("Age Distribution")
plt.xlabel("Age")

plt.subplot(122) # subplot(baris, kolom , letak plot)
plt.pie(gender.values, labels = gender.index, autopct = "%1.2f%%")
plt.title("Gender Distribution")

plt.legend()
plt.tight_layout() # berfungsi untuk memastikan plot tidak tabrakan
plt.show()
```

```python
plt.figure(figsize = (10,9))

plt.subplot(221) # subplot(baris, kolom, letak plot )
plt.plot(age.index, age.values, color = "green", marker = "o", ls = "--")
plt.title("Age Distribution")
plt.xlabel("Age")

plt.subplot(222) # subplot(baris, kolom , letak plot)
plt.pie(gender.values, labels = gender.index, autopct = "%1.2f%%")
plt.title("Gender Distribution")
plt.legend(loc = "upper left")

plt.subplot(223)
plt.bar(age.index, age.values, color = "green")
plt.title("Age Distribution")
plt.xlabel("Age")

plt.subplot(224)
plt.hist(age.index, color = "green", bins = 5)
plt.title("Age Distribution")
plt.xlabel("Nilai")


plt.tight_layout() # berfungsi untuk memastikan plot tidak tabrakan
plt.show()
```

```python
# seaborn

import seaborn as sns

source = data[["sex", "age", "cp", "trestbps", "chol"]]
print(source)
```

```python
# histplot seaborn

sns.histplot(x = "age", data = source, bins = 5)
plt.title("Age Distribution")
plt.show()
```

```python
# scatter seaborn

sns.scatterplot(x ="trestbps", y = "chol", data = source)
plt.title("Heart Disease")
plt.show()
```

```python
# KDE seaborn

sns.kdeplot(x = "trestbps", data = source)
plt.title("Age Distribution")
plt.show()
```

```python
# Barplot

sns.barplot(x = "sex", y = "age", data = source)
plt.title("Bar Plot")
plt.show()
```

```python
# countplot

sns.countplot(x = "sex", data = source)
plt.title("Count Plot")
plt.show()
```

```python
# Boxplot

sns.boxplot(x = "sex", y = "trestbps", data = source)
plt.title("Box Plot")
plt.show()
```

```python
#strip plot

sns.stripplot(x = "sex", y = "trestbps", data = source)
plt.title("Strip Plot")
plt.show()
```

```python
# heatmap

pivot = source.pivot_table(index = "sex", columns = "age", values = "chol")
pivot = pivot.fillna(0)
pivot
```

```python
sns.heatmap(pivot)
plt.title("Heatmap")
plt.show()
```

```python
# hue seaborn
plt.figure(figsize = (10,5))
sns.barplot(x = "sex", y = "chol", data = source, hue = "cp")
plt.title("Bar plot")
plt.show()
```

```python
# subplot (plot yang lebih dari 1)

plt.figure(figsize = (10,5))

plt.subplot(121)
sns.barplot(x = "sex", y = "chol", data = source)

plt.subplot(122)
sns.boxplot(x = "sex", y = "trestbps", data = source)

plt.tight_layout()
plt.show()


```

```python
data2 = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/Cause-of-Death-Indonesia.csv")
print(data2)
```

```python
data2_a = data2.groupby("Year")["Total Deaths"].sum()
print(data2_a)
```

```python
plt.figure(figsize = (10,5))
plt.bar(data2_a.index, data2_a.values, color = "red")
plt.title("Visualisasi")
plt.xlabel("Year")
plt.ylabel("Total Deaths")
plt.show()
```

```python
print(data2_a.idxmax())
```

```python
# data dictionary

data_dictionary = {
    "nama" : ["agung", "elihardo"],
    "alamat" : ["denpasar", "medan"],
    "hobi" : "sepak bola"

}

for i in data_dictionary :
  print("key = ", i ,", value =", data_dictionary[i])
  for z in data_dictionary[i] :
    print(z)
```

```python
jenis_olahraga = {
    "olahraga" : "sepak bola"
}

print("pilih menu")
print("1. tambah")
print("2. hapus")

menu = input("mashukkan menu :")

if menu == "1":
  jenis = input("masukkan jenis olahraga :")
  nama = input("masukkan nama olahraga :")
  jenis_olahraga.update({jenis : [nama]})

elif menu == "2":
  jenis = input("masukkan jenis olahraga :")
  del jenis_olahraga[jenis]

else:
  print("menu tidak tersedia")


print(jenis_olahraga)
```

```python
# fungsi items (memisahkan data key dan value)

data_dictionary = {
    "nama" : ["agung", "elihardo"],
    "alamat" : ["denpasar", "medan"],
    "hobi" : "sepak bola"
}

print(data_dictionary.items())

for a, b in data_dictionary.items():
  print("key = ",a ,",value =",b)
```

```python
# fungsi keys (mengambil data key saja)

data_dictionary = {
    "nama" : ["agung", "elihardo"],
    "alamat" : ["denpasar", "medan"],
    "hobi" : "sepak bola"
}

print(data_dictionary.keys())

for a in data_dictionary.keys():
  print(a)

```

```python
 # fungsi value (mengambil data value saja)

data_dictionary = {
    "nama" : ["agung", "elihardo"],
    "alamat" : ["denpasar", "medan"],
    "hobi" : "sepak bola"
}

print(data_dictionary.values())
```

```python
# fungsi Copy (menyalin data)

data_dicitonary = {
    "nama" : ["agung", "elihardo"],
    "alamat" : ["denpasar", "medan"],
    "hobi" : "sepak bola"
}

data2 = data_dicitonary

data_dicitonary["nama"] = "wahyu"
print(data2)
```

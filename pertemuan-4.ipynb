```python
# slicing = menanmpilkan value dengan rentang index

data = ["satu", "dua", "tiga", "empat", "lima", "enam"]

print(data[0:3]) # start : stop
print(data[0:4:2])    #start : stop : step
```

```python
# slicing negatif
data = ["satu", "dua", "tiga", "empat", "lima", "enam"]


print(data[-1])
print(data[-6 : -2:])
print(data[-6 : -2 : 2])
print(data[::-1])
```

```python
# mengambil value List
data = ["satu", "dua", "tiga", "empat"]

data[1] = "seribu"
data[2:] = ["duaribu", "tigaribu"]
print(data)
```

```python
# menambahkan Value (append , insert)

# append = menambahkan data langsung di bagian akhir

data = ["satu", "dua", "tiga", "empat"]
data.append("lima")
print(data)
```

```python
# menambahkan value sesuai posisi yang diinginkan

data = ["satu", "dua", "tiga", "empat"]
data.insert(1, "halo")
print(data)
```

```python
# menghapus data (del, remove, pop, clear)

# del = menghapus data sesuai index

data = ["satu", "dua", "tiga", "empat"]
del data[1]
print(data)
```

```python
# remove = mengahapus data sesuai value
data = ["satu", "dua", "tiga", "empat"]
data.remove("dua")
print(data)
```

```python
# pop = menghapus data tapi yang dihapus bisa disimpan

data = ["satu", "dua", "tiga", "empat"]
simpan = data.pop(1)
print(data)
print(simpan)
```

```python
# clear = menghapus semuanya

data = ["satu", "dua", "tiga"]
data.clear()
print(data)
```

```python
# list multidimensi

data1 = ["satu", "dua", "tiga"]
data2 = [["satu", "dua", "tiga"], ["empat", "lima", "enam"]]

print(data2[0])
print(data2[0][0])
print(data2[1][2])
```

```python
# loping dalam looping


for i in data2:
  for a in i:
    print(a)
```

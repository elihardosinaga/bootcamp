```python
def tambah():
  try:
    num1 = int(input("masukkan angka 1 :"))
    num2 = int(input("masukkan angka 2 :"))
    print(num1)
    print(num2)
    print("hasil", num1 + num2)
  except ValueError:
    print("value salah")

def kurang():
  try:
    num1 = int(input("masukkan angka 1 :"))
    num2 = int(input("masukkan angka 2 :"))
    print(num1)
    print(num2)
    print("hasil", num1 - num2)
  except ValueError:
    print("value salah")
def kali():
  try:
    num1 = int(input("masukkan angka 1 :"))
    num2 = int(input("masukkan angka 2 :"))
    print(num1)
    print(num2)
    print("hasil", num1 * num2)
  except ValueError:
    print("value salah")

def bagi():
  try:
    num1 = int(input("masukkan angka 1 :"))
    num2 = int(input("masukkan angka 2 :"))
    print(num1)
    print(num2)
    print("hasil", num1 / num2)
  except ValueError:
    print("value salah")

while True:
  print("silahkan memilih menu : ")
  print("1. penjumlahan")
  print("2, pengurangan")
  print("3. perkalian")
  print("4. pembagian")

  kondisi = input("masukkan pilihan anda :")

  if kondisi == "1":
    tambah()
  elif kondisi == "2":
    kurang()
  elif kondisi == "3":
    kali()
  elif kondisi == "4":
    bagi()
  else:
    print("kondisi salah")

  berhenti = input("apakah anda ingin berhenti? (ya/tidak)")
  if berhenti == "ya":
    break
```

```python
# funcition (void, parameter, return )

def tambah1():      #
  num1 = int(input("masukkan angka : "))
  num2 = int(input("masukkan angka : "))
  print(num1 + num2)

def tambah2(num1, num2):    # function parameter = function yang dapat menerimana nilai diluar dirinya sendiri
  hasil = num1 + num2
  print(hasil)

def tambah3(num1, num2):    # function return = function dapat mengembalikan nilai
  hasil = num1 + num2
  return hasil

tambah1()     # Panggil function void
tambah2(10, 20)   # panggil function parameter
result = tambah3(10, 20)    #panggil function return
print(result)
```

```python
# function di dalam function

def persegi_panjang(panjang, lebar):
  luas = panjang * lebar
  return luas

def balok(panjang, lebar, tinggi):
  volume = persegi_panjang(panjang, lebar) * tinggi
  return volume

balok1 = balok(10, 30, 5)
balok2 = balok(20, 40, 6)

print("Volume balok 1 : ", balok1)
print("Volume balok 2 : ", balok2)
```

```python
# function recursive = function yang dapat memanggil dirinya sendiri

def akar(angka): # angka = 5
  if angka == 0:
    return 1
  else:
    return angka * akar(angka - 1)
print(akar(5))
```

```python
def tambah(num1, num2):
  hasil = num1 + num2
  return hasil

def kurang(num1, num2):
  hasil = num1 - num2
  return hasil

def kali(num1, num2):
  hasil = num1 * num2
  return hasil

def bagi(num1, num2):
  hasil = num1 / num2
  return hasil

while True:
  print("silahkan memilih menu : ")
  print("1. penjumlahan")
  print("2, pengurangan")
  print("3. perkalian")
  print("4. pembagian")

  kondisi = input("masukkan pilihan anda :")
  num1 = int(input("masukkan angka 1 :"))
  num2 = int(input("masukkan angka 2 :"))

  if kondisi == "1":
    result = tambah(num1, num2)
    print(result)
  elif kondisi == "2":
    result = kurang(num1,num2)
    print(result)
  elif kondisi == "3":
    result = kali(num1, num2)
    print(result)
  elif kondisi == "4":
    result = bagi(num1, num2)
    print(result)
  else:
    print("value salah")

  berhenti = input("apakah anda ingin berhenti (ya/tidak)")
  if berhenti == "ya":
    break


```

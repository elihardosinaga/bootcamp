```python
# tipe data

kata = "halo 10" # string (kata / kalimat)
angka_int = 50 # int / integer (tipe data untuk bilangan bulat)
angka_float = 10.5 #  (tipe data untuk bilangan desimal)
benar = True # boolean (tipe data untuk benar / salah)
salah = False #

print("Kalimat =", kata)
print("tipe data=", type(kata))
print("angka=", angka_int)
print("tipe data=", type(angka_int))
print("angka=", angka_float)
print("tipe data=", type(angka_float))
print("benar=", benar)
print("tipe data=", type(benar))
print("salah=", salah)
print("tipe data=", type(salah))
```

```python
# konversi data

angka_1 = 10
angka_2 = float(angka_1)

print("nilai=", angka_1)
print("tipe data=", type(angka_1))
print("nilai=", angka_2)
print("tipe data=", type(angka_2))
```

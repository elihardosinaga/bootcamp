```python
# ini adalah Tipe Data

a = "satu"
b = 12
c = 12.4
d = True
e = False


print("a = ", a)
print("a adalah jenis", type(a))
print("b = ", b)
print("b adalah jenis", type(b))
print("c = ", c)
print("c adalah jenis", type(c))
print("d = ", d)
print("d adalah jenis", type(d))
print("e = ", e)
print("e adalah jenis", type(e))
```

```python
# ini contoh dari mengkoversi data dari string (str) ke integer (int)

a = 10
b = float(a)

print(a)
print("ini masuk ke dalam tipe data ", type(a))
print(b)
print("ini masuk ke dalam tipe data ", type(b))
```

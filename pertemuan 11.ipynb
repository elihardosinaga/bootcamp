```python
import pandas as pd

turis_bali = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/bali_tourist.csv")
print(turis_bali)


```

```python
kolom = turis_bali.loc[0:, "January" :].columns
per_bulan_2018 = turis_bali.loc[turis_bali["year"] == 2018, kolom].sum()
per_bulan_2019 = turis_bali.loc[turis_bali["year"] == 2019, kolom].sum()

total_2018 = per_bulan_2018.sum()
rata_2018 = per_bulan_2018.mean()

total_2019 = per_bulan_2019.sum()
rata_2019 = per_bulan_2019.mean()

print("total 2018 :", total_2018)
print("rata-rata 2018 :", rata_2018)

print("total 2019 :", total_2019)
print("rata-rata 2019 :", rata_2019)


```

```python
terbanyak_2018 = per_bulan_2018.idxmax()
terdikit_2018 = per_bulan_2018.idxmin()

terbanyak_2019 = per_bulan_2019.idxmax()
terdikit_2019 = per_bulan_2019.idxmin()

print("bulan terbanyak 2018 :", terbanyak_2018)
print("bulan terdikit 2018 :", terdikit_2018)

print("bulan terbanyak 2019 :", terbanyak_2019)
print("bulan terdikit 2019 :", terdikit_2019)
```

```python
group = turis_bali.groupby("area")[kolom].sum().sum(axis = 1)
print(group.idxmax())
```

```python
group2 = turis_bali.groupby(["area","country"])[kolom].sum().sum(axis = 1)
print(group2.idxmax())
```

```python
tahun = turis_bali.groupby("year")[kolom].sum()
frame = tahun.reset_index()

frame["Max"] = frame[kolom].max(axis = 1)
frame["Min"] = frame[kolom].min(axis = 1)
frame["Total"] = frame[kolom].sum(axis = 1)
frame["Rata-rata"] = frame[kolom].mean(axis = 1)

frame
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

a = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/Cause-of-Death-Indonesia.csv")
print(a)
```

```python
a.info()
```

```python
# soal 1

a_1 = a.groupby("Year")["Total Deaths"].sum()
print(a_1)
```

```python
plt.figure(figsize=(10, 6))
plt.bar(a_1.index, a_1.values, color = "red")
plt.xlabel("Year")
plt.ylabel("Total Deaths")
plt.title("Total Deaths by Year" )
plt.show()
```

```python
print(a_1.idxmax())
```

```python
# soal 2

a_2 = a.loc[a["Type"] == "Bencana Sosial"]
print(a_2)
```

```python
a_3 = a_2.groupby("Cause")["Total Deaths"].sum()
print(a_3)
```

```python
plt.bar(a_3.index, a_3.values, color = "red")
```

```python
print(a_3.idxmax())
```

```python
print(a_3.idxmin())
```

```python
# combining Data

data_2017 = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/user_age_2017.csv")
data_2018 = pd.read_csv("/content/drive/MyDrive/bootcamp data analyst/user_age_2018.csv")

print(data_2017)
print(data_2018)
```

```python
df_combine = pd.concat([data_2017, data_2018])
print(df_combine)
```

```python
df_combine = pd.concat([data_2017, data_2018], ignore_index = True)
print(df_combine)
```

```python
df_combine ["date"] = pd.to_datetime(df_combine ["date"])
print(df_combine.info())
print(df_combine)
```

```python
from datetime import datetime # untuk menghitung jumlah perbedaan hari dari tanggal di data hingga hari ini

tanggal = datetime.now()

df_combine["hari"] = (tanggal - df_combine["date"]).dt.days
print(df_combine)
```

```python
bins = [0, 3000, 4000]
labels = ["tidak lama", "lama"]

df_combine["status"] = pd.cut(df_combine["hari"], bins = bins, labels = labels)
print(df_combine)
```

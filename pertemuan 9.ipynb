```python
import pandas as pd

data = pd.Series([1500, 2000, 3000, 3400, 5500],
                 index = ["a","b","c","d","e"])

print(data)
print(data["a"])
```

```python
print(data > 2000)
```

```python
print(data[data > 2000])
```

```python
print(data[(data > 2000) & (data < 5500)])
```

```python
data_baru = data[(data > 2000) & (data < 5500)]
print(data_baru)
```

```python
print(data.mean())
print(data[data > data.mean()])
```

```python
print(data + 1000)
```

```python
data = data + 1000
print(data)
```

```python
series_baru = pd.Series([1223, 222, 34, 455, 111],
                        index = ["a", "b", "c", "d", "e"])

baru = data + series_baru
print(data)
print(baru)
```

```python
test = data.iloc[0:3] + series_baru.iloc[0:3]
print(test)
```

```python
test = data.iloc[0:6] + series_baru.iloc[0:6]
print(test)
```

```python
# FUNCTION

print(data)
print(data.sum())   # fungsi untuk mencari hasil
print(data.mean())    # fungsi untuk mencari rata rata
print(data ** 2)    # fungsi untuk hasil pangkat
```

```python
print(data)
data = data.astype(float)
print(data)
data = data.astype(int)
print(data)
```

```python
print(data)
data["b"] = 3576
print(data)
```

```python
print(data)
data.iloc[0:3] = [1002, 3245, 5432]
print(data)
data.iloc[0:4] = 1234
print(data)
```

```python
# Dataframe (baris  & kolom)

data_list = [
    [11, 22],
    [32, 45],
    [12, 13],
    [23, 42],
    [12, 34]
]

data_dict = {
    "a" : [11, 32, 12, 23, 12],
    "b" : [22, 45, 13, 42, 34],
    "c" : [12, 12, 12, 12, 12]
}

df_list = pd.DataFrame(data_list, columns = ["a", "b"])
df_dic = pd.DataFrame(data_dict)

print(df_list)
print(df_dic)
```

```python
df_dic.columns = ["kolom_a", "kolom_b", "kolom_c"]
df_dic.index = ["satu", "dua", "tiga", "empat", "lima"]
print(df_dic)
```

```python
test = df_dic["kolom_a"]
print(test) # jadi data series

```

```python
print(df_dic.iloc[0])
print(df_dic.loc["satu"])
```

```python
print(df_dic.iloc[0:3, 0:2])
print(df_dic.loc["satu":"tiga", "kolom_a":"kolom_b"])
```

```python
print(df_dic["kolom_a"])  # otomatis jadi series
print(df_dic[["kolom_a", "kolom_b"]])  # otomatis jadi Datafame
```
